## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd


url = "https://raw.githubusercontent.com/2k25cse2512541-wq/assignment1-flyrank-ml/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)

total_pages = len(df)
declining_pages = len(df[df["trend_direction"] == "down"])
pct_declining = (declining_pages / total_pages) * 100

total_impressions = df["impressions_90d"].sum()
declining_impressions = df[df["trend_direction"] == "down"]["impressions_90d"].sum()
pct_impressions_at_risk = (declining_impressions / total_impressions) * 100

stale_visible = len(df[(df["content_age_days"] >= 180) & (df["impressions_90d"] >= 500)])

print(f"1. Total Pages: {total_pages:,} | Declining Pages: {declining_pages:,} ({pct_declining:.1f}%)")
print(f"2. Total 90d Impressions at Risk: {declining_impressions:,} out of {total_impressions:,} ({pct_impressions_at_risk:.1f}%)")
print(f"3. High-Demand Stale Pages (Age >= 180d, Imps >= 500): {stale_visible:,}")
```
